```python
import numpy as np
from  scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import DSM
```

```python
a = 1
b = 1
c = 1
pi = 1

k = 1
r = 1
C = 1

eps = 1
m = 1 
n = 1
gc = 0.6

BP = 2
R = 2
BM = 2
W = 2

t0 = 0
tf = 10
t_span = (t0, tf)

y0 = np.array([0.5, 0.5, 0.5])
A = np.random.rand()
T = 2
print(A)

param_rand = np.random.rand(11)
```

```python
def ode(t, y, a, b, c, k, pi, eps, m, gc, n, A, T): 
    x = y[0]
    g = y[1]
    d = y[2]

    dx = (a*A*(1-x) - (b + c*T)*x)
    dg = (k*x*(1-g) - pi*g)
    dd = (eps*(m*max(0, g-gc)*(1-d) - n*T*d))

    z = np.zeros(3,)
    z[0] = dx
    z[1] = dg
    z[2] = dd

    return z
```

```python
sol = solve_ivp(ode, t_span, y0, args=(a, b, c, k, pi, eps, m, gc, n, A, T))
```
